# data visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

### BMW yearly trends
     This chart shows the reflecting changes in market conditions and customer demands

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="Year",y="Sales_Volume",estimator="sum",ci=None,marker="*",color="orange",markersize="12")
plt.title("BMW yearly sales Trends(2010-2024)")
plt.xlabel("year")
plt.ylabel("Total sales volume")
plt.grid(True,alpha=0.3)
plt.show()
```

### BMW sales distribution by region
This chart help to compare which region contributes more to BMW's total sales

```python
plt.figure(figsize=(10,6))
ax=sns.barplot(data=df,x="Region",y="Sales_Volume",estimator="sum",ci=None)

plt.title("BMW sales distribution by Region")
plt.xlabel("Region")
plt.ylabel("Total sales volume")
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(
        format(int(p.get_height()),','),
        (p.get_x()+p.get_width()/2.,
         p.get_height()),
        ha='center',va='bottom',
        fontsize=10
    )
plt.show()
```

### BMW top best selling models
This chart helps to identify the customer preference , high demand models ands Models to focus on for marketing and production

```python
top_models=df.groupby("Model")["Sales_Volume"].sum().sort_values(ascending=False).head(9)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=top_models.index,y=top_models.values)
plt.title("BMW Top best selling models")
plt.ylabel("Total sales volume")
plt.xlabel("Model")
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(
        format(int(p.get_height()),','),
        (p.get_x()+p.get_width()/2.,
         p.get_height()),
        ha='center',va='bottom',
        fontsize=10
    )

plt.show()
```

### Hexbin plot of Mileage(KM) and Price(USD)
      This plot shows that as mileage increases the price generally decreases showing a negative relationship between mileage and price

```python
plt.figure(figsize=(10,6))
plt.hexbin(df['Mileage_KM'],df['Price_USD'],gridsize=40)
plt.colorbar(label='Car Count')
plt.title("Mileage vs. Price (density view)")
plt.xlabel("Mileage (KM)")
plt.ylabel("Price (USD)")
plt.show()
```

### Fuel type distribution
This chart analysis shows fairly evenly distributed across different fuel types

```python
plt.figure(figsize=(8,5))
df["Fuel_Type"].value_counts().plot(kind="pie",autopct='%1.1f%%',startangle=90,figsize=(6,6))
plt.title("Fuel Type Distribution", fontsize=14)
plt.ylabel("")
plt.show()
```

### correlation heatmap
    This heatmap shows the correlation between the numeric values in BMW dataset

```python
numeric_df=df.select_dtypes(include=[np.number])
if len(numeric_df.columns)>=4:
    plt.figure(figsize=(5,4))
    corr=numeric_df.corr()
    sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
else:
    print('Not enough numeric columns for a correlation heatmap')
    
```

### barplot for Price vs Engine size 
    This barplot the price differ against the engine size by the fuel type

```python
plt.figure(figsize=(20,6))
sns.barplot(data=df,x="Engine_Size_L",y="Price_USD",hue="Fuel_Type",alpha=0.7)
plt.title("Price Vs. Engine Size(by Fuel Type)",fontsize=14)
plt.xlabel("Engine size (Liters)")
plt.ylabel("Price (USD)")
plt.legend(title="type of fuel")
plt.show()
```
